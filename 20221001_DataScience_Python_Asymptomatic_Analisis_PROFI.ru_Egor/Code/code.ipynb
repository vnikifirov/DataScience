```python
# Calculate the natural log in Python with numpy.log as ln
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.array(range(1, 1024))
y = np.log(x)
plt.plot(x, y)
plt.title('Plotting y=ln(x) with matplotlib')
plt.show()
```

```python
class myNode:
    def __init__(self, data):
        self.data = data
        self.nextNode = None 
a = myNode(1) 
b = myNode(2)
c = myNode(3)

a.nextNode = b
b.nextNode = c 
c.nextNode = a

def countNodes(node):
    counter = 0
    first = node
    while node.nextNode != None:
        counter += 1
        if node.nextNode == first:
            break
        node = node.nextNode
    return counter

num = countNodes(a)
print(num)
```
