```python
import pandas as pd
```

```python
key_rate = pd.read_excel('00_key_rate.xlsx')
```

```python
key_rate['date'][0].day
```

```python
day_points = [1, 5, 10, 15, 20, 25, 28]
```

```python
# df = key_rate.to_frame()
```

```python
key_rate['day'] = key_rate['date']
```

```python
key_rate
```

```python
key_rate['day'] = (key_rate.apply(lambda x: x['day'].day, axis=1))
```

```python
key_rate
```

```python
key_rate_days = key_rate[key_rate['day'].isin(day_points)]
```

```python
key_rate_days = key_rate_days.sort_values(by='date')
```

```python
key_rate_days
```

```python
key_rate_res = key_rate_days[['date', 'rate'] ]
```

```python
key_rate_res.to_csv('00_key_rate.csv', index = False)
```

```python
df = pd.read_excel('04_bas_dep.xlsx')
df['date'] = df['date'].map(lambda x: pd.datetime(x.year, x.month, 1))
df['day'] = df['date']
df['day'] = (df.apply(lambda x: x['day'].day, axis=1))
```

```python
r = pd.date_range(start=df.date.min(), end=df.date.max())
df = df.set_index('date').reindex(r).fillna(method='ffill').rename_axis('date').reset_index()
df['day'] = df['date']
df['day'] = (df.apply(lambda x: x['day'].day, axis=1))
df = df[df['day'].isin(day_points)]
df = df.sort_values(by='date')
```

```python
df
```
