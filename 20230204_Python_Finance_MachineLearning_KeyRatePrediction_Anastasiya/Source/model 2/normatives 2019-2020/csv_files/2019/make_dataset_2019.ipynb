```python
def get_df_from_file(file):
    with open(file, encoding='utf-8') as input_file:
        d = dict()
        for line in input_file.readlines()[1:]:
            #print(line.split(';'))
            s = line.split(';')
            d[s[0]] = dict()
        #print(d)
    with open(file, encoding='utf-8') as input_file:
        for line in input_file.readlines()[1:]:
            s = line.split(';')
            #print(s[0], s[1], s[2])
            d[s[0]][s[1]] = float(s[2].strip().replace(',', '.'))
        #print(d)
    df = pd.DataFrame.from_dict(d, orient='index')
    return df
```
